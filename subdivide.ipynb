```python
import math

class Scale(object):

    def __init__(self, scale, extend=False, low=(100,100,100), high=(100,100,100)):
        self.scale = scale
        self.extend = extend
        self.low = low
        self.high = high

    def _rgbs(self):
        num_colors = len(self.scale) - 1

        r, g, b = 0.0, 0.0, 1.0

        if num_colors < 2:
            return [(round(r * 255, 0), round(g * 255, 0), round(b * 255, 0))]

        incr = (4.0 / (float(num_colors) - 1.0))
        tol = 0.0001

        rgb_vals = []

        # Traverse along green, holding b, r
        while g < 1.0:
            rgb_vals.append((round(r * 255, 0), round(g * 255, 0), round(b * 255, 0)))
            g += incr

        b, g = 2.0 - g, 1.0

        # Traverse along b, holding g, r
        while b > 0.0:
            rgb_vals.append((round(r * 255, 0), round(g * 255, 0), round(b * 255, 0)))
            b -= incr

        r, b = 0 - b, 0.0

        # Traverse along r, holding g, b
        while r < 1.0:
            rgb_vals.append((round(r * 255, 0), round(g * 255, 0), round(b * 255, 0)))
            r += incr

        g, r = 2 - r, 1.0

        # Traverse along g again
        while g >= -1 * math.fabs(tol):
            rgb_vals.append((round(r * 255, 0), round(g * 255, 0), round(b * 255, 0)))
            g -= incr

        return rgb_vals

    def rgb(self, val):

        def _iter_scale(scale):
            i = iter(scale)
            a = i.next()

            for b in i:
                yield a, b
                a = b

        rgbs = self._rgbs()

        if val < self.scale[0]:

            if self.extend:
                return rgbs

            else:
                return self.low

        for (low, high), rgb in zip(_iter_scale(self.scale), rgbs):

            if low <= val and val <= high:
                return rgb

        if self.extend:
            return rgbs[-1]

        else:
            return self.high

def sub_div3(face, scale, attr, max_divs):

    # If the maximum number of divisions allows is 0, then just call it quits and return
    if max_divs == 0:
        return [face]

    # Get each node of the triangle
    n0 = face[0]
    n1 = face[1]
    n2 = face[2]

    if attr not in n0:
        raise KeyError("node 0 missing node attribute {0}".format(attr))

    if attr not in n1:
        raise KeyError("node 1 missing node attribute {0}".format(attr))

    if attr not in n2:
        raise KeyError("node 2 missing node attribute {0}".format(attr))

    r0, g0, b0 = scale.rgb(n0[attr])
    r1, g1, b1 = scale.rgb(n1[attr])
    r2, g2, b2 = scale.rgb(n2[attr])

    # No need to divide, just return
    if r0 == r1 and r1 == r2 and g0 == g1 and g1 == g2 and b0 == b1 and b1 == b2:
        return [face]

    # Face is not all the same, and there are available divides, so get to it
    n01 = {}
    n12 = {}
    n20 = {}

    for item in n0.keys():
        n01[item] = 0.50 * (n1[item] + n0[item])
        n12[item] = 0.50 * (n2[item] + n1[item])
        n20[item] = 0.50 * (n0[item] + n2[item])

    f0 = [n0,  n01, n20]
    f1 = [n01, n1,  n12]
    f2 = [n20, n12, n2]
    f3 = [n01, n12, n20]

    sub_faces = [f0, f1, f2, f3]
    faces = []

    for sub_face in sub_faces:
        faces.extend(sub_div3(sub_face, scale, attr, max_divs - 1))

    return faces
```
